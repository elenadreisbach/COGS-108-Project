```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
df_2009 = pd.read_excel('unemployment_2009.xlsx', skiprows=5)
df_2009.drop(['Unnamed: 0' , 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 5'], axis = 1, inplace=True)
df_2009.columns = ['State', 'Year', 'Labor Force', 'Employed', 'Unemployed', 'Unemployment Rate (%)']
df_2009['State']=df_2009['State'].str[-2:]
df_2009 = df_2009.set_index(['State'])
```

```python
df_2009=df_2009.groupby(df_2009.index).sum()
df_2009['Unemployment Rate (%)'] = df_2009['Labor Force'] / df_2009['Unemployed'] 
df_2009['Year']=2009
df_2009=df_2009[:-1]
```

```python
#df_2009
```

```python
df_2010 = pd.read_excel('unemployment_2010.xlsx', skiprows=5)
df_2010.drop(['Unnamed: 0' , 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 5'], axis = 1, inplace=True)
df_2010.columns = ['State', 'Year', 'Labor Force', 'Employed', 'Unemployed', 'Unemployment Rate (%)']
df_2010['State']=df_2010['State'].str[-2:]
df_2010 = df_2010.set_index(['State'])
df_2010=df_2010.groupby(df_2010.index).sum()
df_2010['Unemployment Rate (%)'] = df_2010['Labor Force'] / df_2010['Unemployed'] 
df_2010['Year']=2010
df_2010=df_2010[:-1]
```

```python
#df_2010
```

```python
df_2011 = pd.read_excel('unemployment_2011.xlsx', skiprows=5)
df_2011.drop(['Unnamed: 0' , 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 5'], axis = 1, inplace=True)
df_2011.columns = ['State', 'Year', 'Labor Force', 'Employed', 'Unemployed', 'Unemployment Rate (%)']
df_2011['State']=df_2011['State'].str[-2:]
df_2011 = df_2011.set_index(['State'])
df_2011=df_2011.groupby(df_2011.index).sum()
df_2011['Unemployment Rate (%)'] = df_2011['Labor Force'] / df_2011['Unemployed'] 
df_2011['Year']=2011
df_2011=df_2011[:-1]
```

```python
#df_2011
```

```python
df_unemployment_by_state=pd.concat([df_2009, df_2010, df_2011], axis=1)
df_unemployment_by_state.to_json('df_unemployment_by_state.json', orient='values')
```
