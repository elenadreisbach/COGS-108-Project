```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def clean(filename):
    df = pd.read_excel(filename, skiprows=5)
    df.drop(['Unnamed: 0' , 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 5', 'Unnamed: 9'], axis = 1, inplace=True)
    df.columns = ['state', 'year', 'labor force', 'employed', 'unemployed']
    year = df.year[0]
    df['state']=df['state'].str[-2:]
    #df = df.set_index(['State'])
    df = df[:-3] # drop last three rows that were noise
    if(df['labor force'].dtype != 'float64'):
        df = df[df['labor force'] != 'N.A.']
    df[['labor force','employed','unemployed']] = df[['labor force','employed','unemployed']].apply(pd.to_numeric)
    df = df.groupby(df.state.values).sum()
    df['unemployment rate'] = df['unemployed'] / df['labor force'] * 100
    df.year = int(year)
    df.reset_index(level=0, inplace=True)
    df['state'] = df['index']
    df = df.drop(['index'], axis = 1)
   
    
    return df
```
